# Recursos lxicos en NLTK

NLTK nos proporciona otro tipo de corpora que son en realidad "lexicon resources" o recursos lxicos o ms simplemente *lexicons*. En su forma ms simple, estos lexicons no son otra cosa que listas de palabras, algunos ordenados, otros etiquetados.



Vamos a ver algunos ejemplos interesantes de los lexicons que nos ofrece NLTK

## Listas de palabras

En su forma ms simple, NLTK nos ofrece algunas listas de palabras en formato plano. En el corpus "words" tenemos una lista de palabras en ingls.

Podemos usarlas para encontrar algunas palabras mal escritas o simplemente palabras inusuales que estn presentes en un texto.

```python
import nltk
```

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words('en'))
    unusual = text_vocab - english_vocab
    return sorted(unusual)
```

```python
#veamos un ejemplo
# qu palabras poco usuales hay en MobyDick
import nltk
unusual_words(nltk.corpus.gutenberg.words('melville-moby_dick.txt'))
print(len(unusual_words(nltk.corpus.gutenberg.words('melville-moby_dick.txt'))))
```

Un lexicon muy interesante son los diccionarios de **stop-words**


```python
#qu encontramos en el lexicon stop-words
for file in nltk.corpus.stopwords.fileids():
    print(file)
```

```python
#qu stop-words en espaol encontramos??
spanish_stop = nltk.corpus.stopwords.words('spanish')
```

```python
len(spanish_stop)
```

```python
for w in spanish_stop:
    print(w)
```

Otro lexicon interesante es "**swadesh**" que contiene listas comparativas de palabras comunes en diferentes idiomas

Ms de cerca:

```python
for file in nltk.corpus.swadesh.fileids():
    print(file)
```

```python
#Comparamos francs (fr) y espaol (es)
es_fr = nltk.corpus.swadesh.entries(['es','fr'])
es_fr
```

```python
#creamos un diccionario para traducir entre uno y otro
translator = dict(es_fr)
translator['yo']
translator['sangre']
```

```python
#comparamos la misma palabra en diferentes lenguajes
languages = ['en', 'de', 'nl', 'es', 'fr', 'pt', 'it']
for i in [139, 140, 141, 142]:
    print(nltk.corpus.swadesh.entries(languages)[i])
```
