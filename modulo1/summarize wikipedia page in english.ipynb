### Limpieza

```python
textolimpio = texto.lower()
textolimpio
```

```python
textolimpio = contractions.fix(textolimpio)
textolimpio
```

```python
# usamos exrpresiones regulares para eliminar subtitulos de la pagina

textolimpio = re.sub(r'={3,}.*?={3,}\n', '', textolimpio)
textolimpio
```

```python
# para fines de analisis vamos a eliminar todo lo que no sean palabras
textolimpio = re.sub(r"\W"," ", textolimpio)
textolimpio =re.sub(r"\d", " " , textolimpio)
textolimpio = re.sub(r'\s+'," ", textolimpio)
textolimpio
```

```python
stop_words = nltk.corpus.stopwords.words("english")
stop_words
```

```python
#texto = texto.lower()
sentences = nltk.sent_tokenize(texto)
```

```python
#Tabla de frecuencias en diccionario
cuentapalabras = {}
for word in nltk.word_tokenize(textolimpio):
    if word not in stop_words:
        if word not in cuentapalabras.keys():
            cuentapalabras[word] = 1
        else:
            cuentapalabras[word] +=1
```

```python
# generamos frecuencias ponderadas
max_count = max(cuentapalabras.values())
for key in cuentapalabras.keys():
    cuentapalabras[key] = cuentapalabras[key] / max_count
```

```python
# frecuencia ponderada para calcular score de cada frase
scorefrases = {}
for sentence in sentences:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in cuentapalabras.keys():
            if len(sentence.split(" ")) < 25:
                if sentence not in scorefrases.keys():
                    scorefrases[sentence] = cuentapalabras[word]
                else:
                    scorefrases[sentence] += cuentapalabras[word]
```

```python
scorefrases
```

```python
# obtenemos las 5 frases con el score mas alto 
primerasfrases = heapq.nlargest(5,scorefrases, key = scorefrases.get)
primerasfrases
```

```python
for frase in primerasfrases:
    print(frase)
```
